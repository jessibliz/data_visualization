This data set contains 113,937 loans with 81 variables on each loan,59 Columns are Numerical  while 22 Columns are Categoricel Variables 


````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
prosperloan = pd.read_csv('prosperLoanData.csv')
````

 ## (A PLOT OF LoanOriginalAmount and MonthlyLoanPayment)

It showed the highest LoanOriginalAmount of 35000 has a MonthlyLoanPayment between 0-1000


````python
#  A plot of LoanOriginalAmount and MonthlyLoanPayment
plt.figure(figsize = [8,8])
plt.scatter(data = prosperloan, x = 'MonthlyLoanPayment', y = 'LoanOriginalAmount', alpha = 1/10)
plt.xlabel('MonthlyLoanPayment')
plt.ylabel('LoanOriginalAmount')
plt.title('LoanOriginalAmount and MonthlyLoanPayment Distribution')
plt.show()
````

## (MonthlyLoanPayment vs OnTimeProsperPayments vs ProsperScore)

The distribution showed a positive corellation between OnTimeProsperPayment and MonthlyLoanPayment as the ProsperScore increases.



````python
plt.scatter(data = prosperloan, x = 'LoanOriginalAmount', y = 'MonthlyLoanPayment', c = 'ProsperScore', cmap = 'inferno_r')
plt.colorbar(label = 'ProsperScore')
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('MonthlyLoanPayment')
plt.title('MonthlyLoanPayment,OnTimeProsperPayments and ProsperScore');
````

## (Distribution of OnTimeProsperPayments)

The distribution of the OnTimeProsperPayment showed a right sudden change in distribution with long tail on a standard scale. 




````python
# start with a standard-scaled plot of OnTimeProsperPayment
binsize = 4
bins = np.arange(0,prosperloan['OnTimeProsperPayments'].max()+binsize, binsize)
plt.figure(figsize=[8,5])
plt.hist(data = prosperloan, x = 'OnTimeProsperPayments', bins = bins)
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('Number of OnTimePayment made')
plt.title('OnTimeProsperPayments on Standard scale');
           
````
